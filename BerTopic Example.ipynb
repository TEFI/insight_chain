```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
import pandas as pd

# 1. List of generated outputs or prompt completions
texts = [
    "The cat is on the mat.",
    "Dogs are great pets.",
    "Transformers are powerful models.",
    "My dog loves to play.",
    "BERT is a transformer-based model.",
    "I enjoy training large language models.",
    "Cats often sleep during the day.",
    "The weather today is sunny and warm.",
    "Reinforcement learning improves AI decision-making.",
    "Pandas are cute but dangerous."
]

# 2. Initialize SBERT to compute text embeddings
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# 3. Create the BERTopic model using the SBERT embeddings
topic_model = BERTopic(embedding_model=embedding_model)

# 4. Fit the model and extract topics and their probabilities
topics, probs = topic_model.fit_transform(texts)

# 5. Visualize and export the interactive plot as HTML
fig = topic_model.visualize_documents(texts)
fig.write_html("bertopic_clusters.html")
print("✅ Visualization saved to: bertopic_clusters.html")
```
